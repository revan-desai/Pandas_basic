```python
import pandas as pd

ipldata=pd.read_csv('/content/iplmatches.csv')

ipldata.head().sort_values(['city','date'],ascending=[True,False])

                     # this is also correct ipldata.head().sort_values(by=['city','date'],ascending=[True,False])
```

## In how many cities ipl is played?

```python
print(ipldata['city'].nunique())

print(ipldata.drop_duplicates(['city']).shape[0])    # i think this is also counting nan (last value)....hence giving 30 instead 31

ipldata['city'].unique()
```

```python
ipldata['city'].fillna('Bangalore',inplace=True)     # we have replaced null(nan) value in city column with bangalore


print(ipldata['city'].nunique())

print(ipldata['city'].unique())

print(ipldata.drop_duplicates(['city']).shape[0])    # now we are getting 30 instead of 31

```

```python
finaldata=ipldata.drop_duplicates(['season'],keep='last')
     # keep last --meaning last record of that season... there are many matches played in particular season or year

finaldata[['season','date','winner']].sort_values(['season'])  # sorting done by season not by values

# only final matches
```
