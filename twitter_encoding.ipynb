```python
import pandas as pd

df=pd.read_csv('/content/data.csv')
```

```python
df.shape
```

```python
df.sample(5)
```

```python
df=df.drop_duplicates()
```

```python
df.shape
```

```python
df=df[['Tweets','Feeling']]
```

```python
df.sample(5)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

df['Feeling']=le.fit_transform(df['Feeling'])
```

```python
df.sample(10)
```

```python
df['Tweets'] = df['Tweets'].astype('category')
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe=OneHotEncoder()

ohe_data = pd.DataFrame(ohe.fit_transform(df[['Tweets']]).toarray())
```

```python
df=df.join(ohe_data)
```

```python
df=df.drop(columns=(['Tweets']),axis=1)
```
