```python
import pandas as pd

df=pd.read_csv("/content/student.csv")

print(df[0:10])                            #observe data visualization we done here [0:10] means first 10records from 0th index

print("***********************************")

df[0:10].sort_values(["name"],axis=0,ascending=True,na_position='last')

```
