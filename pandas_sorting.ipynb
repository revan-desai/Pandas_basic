## Group By () function in pandas

```python
# importing pandas module
import pandas as pd

# Define a dictionary containing employee data
data1 = {'First Name':['Jai', 'Anuj', 'Jai', 'Princi',
				'Gaurav', 'Anuj', 'Princi', 'Abhi'],
		'Age':[27, 24, 22, 32,
			33, 36, 27, 32],
		'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj',
				'Jaunpur', 'Kanpur', 'Allahabad', 'Aligarh'],
		'Qualification':['Msc', 'MA', 'MCA', 'Phd',
						'B.Tech', 'B.com', 'Msc', 'MA']}

df=pd.DataFrame(data1,index=['Jai sharma', 'Anuj Patel', 'Jai singh', 'Princi',
				'Gaurav', 'Anuj Kumar', 'Princi Mehra', 'Abhi'])

print(df)

df1=df.groupby(['Address']).groups

df1



```

```python
df2=df.groupby(['First Name','Qualification']).groups

df2
```

```python
df3=df.groupby(['First Name','Age']).groups

df3
```

```python
df4=df.groupby(['First Name'])['Age'].sum()
df4
```

Note: Aggregations:
count() – Number of non-null observations

sum() – Sum of values

mean() – Mean of values

median() – Arithmetic median of values

min() – Minimum

max() – Maximum

mode() – Mode

std() – Standard deviation

```python
import numpy as np

grp1=df.groupby(['First Name'])

grp1['Age'].aggregate(np.sum)

grp1['Age'].aggregate([np.sum,np.mean,np.std])   # make a habit of putting multiple arguments in square brackets


# std means standard deviation

```
