```python
import pandas as pd

ipldata=pd.read_csv('/content/iplmatches.csv')

delivery=pd.read_csv('/content/deliveries.csv')
```

```python
new=delivery.merge(ipldata,left_on='match_id',right_on='id')     #we have column name "match_id" in delivery data set and "id" in ipldata dataset

new.head(1)
```

# top run scorer season wise

```python
new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates('season',keep='first').sort_values('season')
```

```python
new1=ipldata.merge(delivery,left_on='id',right_on='match_id')

new1.head(1)
```

```python
new1.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates('season',keep='first').sort_values('season')
```
