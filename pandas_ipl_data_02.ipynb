# masking in pandas

```python
import pandas as pd

import matplotlib.pyplot as plt

ipldata=pd.read_csv(r'/content/iplmatches.csv')

ipldata.head(2)
```

```python
ipldata['city']=='Pune'
```

```python
mask=ipldata['city']=='Pune'

ipldata[mask]
```

```python

def get_city_match_count(city):
  mask=ipldata['city']==city
  return ipldata[mask].shape[0]        # shape ....return tuple... (row,columns)


get_city_match_count('Mumbai')
```

## total matches played by mumbai indians as a team1 and team2

```python
ipldata['team1'].value_counts()+ipldata['team2'].value_counts()  # total matches played by mumbai indians or any other team
```
