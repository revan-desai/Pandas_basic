```python
import pandas as pd

dict2={'cars':['honda','hyundai','tata'],'year':['2011','2020','2022']}

df=pd.DataFrame(dict2)

df1=df.rename(columns={'cars':'4wheeler','year':'manuYear'},inplace=False)
print(df)
print("**************************************")
df1
```
