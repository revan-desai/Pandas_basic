```python
import pandas as pd

import seaborn as sns

data=pd.read_csv('/content/display.csv')

data.head(3)
```

## 1.replace spaces in all column names with underscore

```python
data.columns=data.columns.str.replace(' ','_')

data.head(3)
```

## 2. to rename columns names

```python
data.rename(columns={'South_American':'South_USA'},inplace=True)

data.head(3)
```

## to rename content of columns

```python
data['Delivery']=data['Delivery'].str.replace('Yes','Yes(within 3Km)')

data.head(3)
```
